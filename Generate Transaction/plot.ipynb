```python
import numpy as np
import matplotlib.pyplot as plt
```

### EXP 1 : Time vs numTrans

```python
logFile = '/home/darpan/Desktop/Desktop/8thSem/CS5280/cctsProject/Generate Transaction/Exp 1/logFile.txt'
with open(logFile, 'r') as f:
    lines = f.readlines()
```

```python
numTrans = np.array([1000, 2000, 3000, 4000, 5000])
dag = []
dagCreationTime = []
blockPilot = []
blockPilotAborts = []
for i, line in enumerate(lines):
    if 'dag' in line:
        temp = 0.0
        dct = 0.0
        for j in range(i + 1, i + 6):
            temp += float(lines[j].split()[0])
            dct += float(lines[j].split()[1])
        dag.append(temp / 5)
        dagCreationTime.append(dct / 5)
    if 'blockpilot' in line:
        temp = 0.0
        aborts = 0
        for j in range(i + 1, i + 6):
            temp += float(lines[j].split()[0])
            aborts += float(lines[j].split()[1])
        blockPilotAborts.append(aborts / 5)
        blockPilot.append(temp / 5)
dag = np.array(dag)/1000
blockPilot = np.array(blockPilot)/1000
dagCreationTime = np.array(dagCreationTime)/1000

plt.plot(numTrans, dag, label='DAG', marker='o')
plt.plot(numTrans, blockPilot, label='Block Pilot', marker='o')
plt.xlabel('Number of Transactions')
plt.ylabel('Time (milliseconds)')
plt.title('Time taken vs Number of Transactions')
plt.legend()
plt.grid()
plt.savefig('numTrans_vs_Time.png', dpi=300)
plt.show()

# Plotting the aborts make bar graph and line graph
x = np.arange(len(numTrans))
width = 0.35
plt.bar(x , blockPilotAborts, width, label='Block Pilot Aborts')
for i in range(len(numTrans)):
    plt.text(x[i], blockPilotAborts[i] + 0.1, str(int(blockPilotAborts[i])), ha='center', va='bottom')
plt.xlabel('Number of Transactions')
plt.ylabel('Number of Aborts')
plt.title('Number of Aborts vs Number of Transactions')
plt.xticks(x, numTrans)
plt.savefig('numTrans_vs_Aborts.png', dpi=300)
plt.show()

print(f"dag: {dag}")
print(f"blockPilot: {blockPilot}")
print(f"blockPilotAborts: {blockPilotAborts}")

# Plotting the DAG creation time
plt.bar(x, dagCreationTime, width, label='DAG Creation Time')
for i in range(len(numTrans)):
    plt.text(x[i], dagCreationTime[i] + 0.1, str(int(dagCreationTime[i])), ha='center', va='bottom')
plt.xlabel('Number of Transactions')
plt.ylabel('DAG Creation Time (milliseconds)')
plt.title('DAG Creation Time vs Number of Transactions')
plt.xticks(x, numTrans)
plt.savefig('numTrans_vs_DAGCreationTime.png', dpi=300)
plt.show()
```
