```python
import numpy as np
import matplotlib.pyplot as plt
```

### EXP 1 : Time vs numTrans

```python
logFile = '/home/darpan/Desktop/Desktop/8thSem/CS5280/cctsProject/Generate Transaction/Exp 1/logFile.txt'
with open(logFile, 'r') as f:
    lines = f.readlines()
```

```python
numTrans = np.array([1000, 2000, 3000, 4000, 5000])
dag = []
blockPilot = []
for i, line in enumerate(lines):
    if 'dag' in line:
        temp = 0.0
        for j in range(i + 1, i + 6):
            temp += float(lines[j].split()[0])
        dag.append(temp / 5)
    if 'blockpilot' in line:
        temp = 0.0
        for j in range(i + 1, i + 6):
            temp += float(lines[j].split()[0])
        blockPilot.append(temp / 5)
dag = np.array(dag)
blockPilot = np.array(blockPilot)

plt.plot(numTrans, dag, label='DAG', marker='o')
plt.plot(numTrans, blockPilot, label='Block Pilot', marker='o')
plt.xlabel('Number of Transactions')
plt.ylabel('Time (microseconds)')
plt.title('Time taken vs Number of Transactions')
plt.legend()
plt.grid()
plt.savefig('numTrans_vs_Time.png', dpi=300)
plt.show()
```
